```python
import os 
path = r"E:\MIDRC"
folders = set([name for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))])
print("Count:", len(folders))
```

```python
folders_manifest = set({item["file_name"].split("/")[0] for item in data})
print(folders_manifest)
print("Count:", len(folders_manifest))
```

```python
from collections import Counter

prefixes = [item["file_name"].split("/")[0] for item in data]
counts = Counter(prefixes)

duplicates = {folder: count for folder, count in counts.items() if count > 1}

print("Repeated folders:")
for folder, count in duplicates.items():
    print(folder, "→", count, "times")

print("\nTotal repeated folders:", len(duplicates))
```

```python
root_folder = r"E:\MIDRC"
all_files = []

for folder, subfolders, files in os.walk(root_folder):
    for f in files:
        full_path = os.path.join(folder, f)
        all_files.append(full_path)
        print(full_path)

print("\nTOTAL archivos encontrados:", len(all_files))
```
